```python
# changing datatypes

df[['Spotify Streams',
       'Spotify Playlist Count', 'Spotify Playlist Reach','YouTube Views', 'YouTube Likes', 'TikTok Posts',
       'TikTok Likes', 'TikTok Views','YouTube Playlist Reach','AirPlay Spins', 'SiriusXM Spins','Deezer Playlist Reach','Pandora Streams', 'Pandora Track Stations',
       'Soundcloud Streams', 'Shazam Counts']]=df[['Spotify Streams',
       'Spotify Playlist Count', 'Spotify Playlist Reach','YouTube Views', 'YouTube Likes', 'TikTok Posts',
       'TikTok Likes', 'TikTok Views','YouTube Playlist Reach','AirPlay Spins', 'SiriusXM Spins','Deezer Playlist Reach','Pandora Streams', 'Pandora Track Stations',
       'Soundcloud Streams', 'Shazam Counts']].replace(',',"",regex=True).astype("Int64")

df["Release Date"]=pd.to_datetime(df["Release Date"])

df["All Time Rank"]=df["All Time Rank"].str.replace(",","").astype("int")


```

```python
df.columns
```

```python
# change column names. I prefer column names without spaces
df=df.rename(columns=lambda x:x.replace(" ","_"))
```

```python
# remove any redundant columns
df.isna().sum()
# we see that TIDAL_Popularity is completely empty.
df=df.dropna(axis=1,how="all").copy()
```

```python
df[df.duplicated()]
# there are two duplicate rows. Dropping them.
df=df.drop_duplicates().copy()
```

```python
# Artist Column has few NA values. Will replace it with "Uknown"
df=df.fillna(value={"Artist":"Unknown"}).copy()
```

```python
# There are some Non-ASCII characters in Artist and Album_Name. I will remove them and replace any empty or values having only space by "Unknown"

b=df[["Artist","Album_Name","Track"]].replace(r"[^\x00-\x7F]+","",regex=True)
df[["Artist","Album_Name","Track"]]=b.map(lambda x:"unknown" if (x==" ") | (x=="") else x ).copy()
```

```python
# what is the count of youtube songs with at least a billion views since it was released

billion_views_youtube=df.query('YouTube_Views>=10**9').groupby("Release_Date")["YouTube_Views"].count().reset_index()
plt.figure(figsize=(20,10))
y_views=sns.lineplot(data=billion_views_youtube,x="Release_Date",y="YouTube_Views")
plt.show()

```

```python
# between 2011 and 2020?
high_views_count=billion_views_youtube.query("Release_Date.dt.year>=2011 & Release_Date.dt.year<=2020 ").reset_index().drop("index", axis=1)
high_views_count.rename(columns={"YouTube_Views":"Count of YouTube Video Songs >= 1,000,000,000"},inplace=True)
plt.figure(figsize=(20,10))
sns.lineplot(data=high_views_count,x="Release_Date",y="Count of YouTube Video Songs >= 1,000,000,000")

plt.show()
```

```python
# How about Spotify streams at least a billion streams since it was released on spotify
billion_streams_spotify=df.query('Spotify_Streams>=10**9').groupby("Release_Date") \
    ["Spotify_Streams"].count().reset_index()
plt.figure(figsize=(20,10))
s_streams=sns.lineplot(data=billion_streams_spotify,x="Release_Date",y="Spotify_Streams")
plt.show()
```

```python
plt.figure(figsize=(20,10))
y_views=sns.lineplot(data=billion_views_youtube,x="Release_Date",y="YouTube_Views")
s_streams=sns.lineplot(data=billion_streams_spotify,x="Release_Date",y="Spotify_Streams")
plt.show()
```

```python
# the linplots show similar counts over the entire period. Lets check their counts correlation.

billion_views_youtube.join(billion_streams_spotify.set_index(["Release_Date"]),on="Release_Date")[["YouTube_Views","Spotify_Streams"]].dropna().corr()

# Pretty high
```

```python

# Top n artists by on different platforms' parameters over different numeric limits
def top_n_artists(n,platform,num):
    top_artists=df.query(platform+">="+str(num))\
    .groupby("Artist") \
    [platform].count().reset_index().sort_values(by=platform,ascending=False)
    return top_artists.head(n)

plt.figure(figsize=(20,10))
s="Spotify_Streams"
sns.barplot(data=top_n_artists(20,s,1000000000),x=s,y="Artist")
```

```python
# Top three months in each year for the last 10 years with respect to number of counts 
dates=pd.concat([df.Release_Date.dt.year.rename("Year"),df.Release_Date.dt.month.rename("Month")],join="inner",axis=1) 

grouped=dates.groupby(["Year","Month"]).size().reset_index(name="count")

grouped["Rank"]=grouped.groupby(["Year"])['count'].rank(method="dense",ascending=False)
pt=grouped.query('Rank<=3 & Year>=2014').head(100)
plt.figure(figsize=(20,10))
sns.barplot(data=pt,x="Year",y="count",hue="Month")


```
